### Loading the Data into Python

```python
import pandas as pd

# To Load the CSV file into a Pandas DataFrame
df = pd.read_csv('c:/Abhi/Carrer/Learn/Guvi/DS/Visual_St/VS_Code/orders.csv')


df.head()
```

#### Checking for missing values

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Get data types of each column
print(df.dtypes)
```

```python
# Replace missing numerical values with 0
df.fillna(0, inplace=True)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Renaming the column 
df.columns = [col.lower().replace(" ", "_") for col in df.columns]
df.head(5)
```

```python
# Trimming spaces from all string columns
for col in df.select_dtypes(include='object').columns:  # Select only string columns
    df[col] = df[col].map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Adding New column 
# Calculate discount
df['discount'] = df['list_price'] * (df['discount_percent'] / 100)

# Calculate sale price
df['sale_price'] = df['list_price'] - df['discount']

# Calculate profit
df['profit'] = (df['sale_price'] - df['cost_price']) * df['quantity']

# Preview the updated DataFrame
print(df.head())
```

```python
# converting the data frame into a CSV file
df.to_csv('SalesInsights.csv',index=False)
```

### Insterting the Datas in SQL

```python
# Spliting the table in two and creating schema for table orders table and storing data
mycursor.execute("""CREATE TABLE Orders (
    order_id INT PRIMARY KEY,
    order_date DATE,
    ship_mode VARCHAR(50),
    segment VARCHAR(50),
    country VARCHAR(50),
    city VARCHAR(50),
    state VARCHAR(50),
    postal_code INT,
    region VARCHAR(50),
    category VARCHAR(50),
    sub_category VARCHAR(50)
)""")

# Extracting relevant columns for the Orders table
orders_data = df[['order_id', 'order_date', 'ship_mode', 'segment', 'country', 
                  'city', 'state', 'postal_code', 'region', 'category', 'sub_category']]

# Inserting the orders into the Orders table
for index, row in orders_data.iterrows():
    sql = """
    INSERT INTO Orders (order_id, order_date, ship_mode, segment, country, city, state, postal_code, region, category, sub_category)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    mycursor.execute(sql, tuple(row))
```

```python
mycursor.execute("SELECT * FROM orders")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
# creating schema for table products table and storing data
mycursor.execute("""CREATE TABLE Products (
    order_id INT,
    product_id VARCHAR(50),
    cost_price DECIMAL(10, 2),
    list_price DECIMAL(10, 2),
    quantity INT,
    discount_percent DECIMAL(5, 2),
    discount DECIMAL(10, 2),
    sale_price DECIMAL(10, 2),
    profit DECIMAL(10, 2),
    PRIMARY KEY (order_id, product_id),
    FOREIGN KEY (order_id) REFERENCES Orders(order_id)
);""")

# Extracting relevant columns for the Products table
products_data = df[['order_id', 'product_id', 'cost_price', 'list_price', 'quantity', 
                    'discount_percent', 'discount', 'sale_price', 'profit']]

# Inserting the product data into the Products table
for index, row in products_data.iterrows():
    sql = """
    INSERT INTO Products (order_id, product_id, cost_price, list_price, quantity, discount_percent, discount, sale_price, profit)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    mycursor.execute(sql, tuple(row))
```

```python
mycursor.execute("SELECT * FROM Products")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
